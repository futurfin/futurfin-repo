```python
def obfuscate(byt):
    # Use same function in both directions.  Input and output are bytes
    # objects.
    mask = b'trap'
    lmask = len(mask)
    return bytes(c ^ mask[i % lmask] for i, c in enumerate(byt))

def test(s):
    data = obfuscate(s.encode())
    print(len(s), len(data), data)
    newdata = obfuscate(data).decode()
    print(newdata)
    
s = 'maxy'
data = obfuscate(s.encode())

print (data)

test(s)
```
